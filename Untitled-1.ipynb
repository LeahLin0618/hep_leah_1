```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
G = 1 / (1.22e19)**2
g_sm = 102.6
M_P = (8 * np.pi * G)**(-0.5)
t = np.logspace(1, 1e17, 1000)  # time in seconds

# Function definitions
def Gamma_PBH(k, M_MBE):
    return (G * g_sm / (7680 * np.pi) * 2**k * (3 + 2*k) * M_P * (M_P / M_MBE)**(3 + 2*k))

def tau_PBH(k, M_MBE):
    return 1 / Gamma_PBH(k, M_MBE)

def M_MBE_evolve(t, tau_PBH, Gamma_PBH_k, M_MBE, k, q=0.5):  # q is an arbitrary small constant
    factor = (1 - Gamma_PBH_k * (t - tau_PBH * (1 - q**3)))**(1 / (3 + 2 * k))
    return M_MBE * factor

# Initial parameters and plot setup
M_MBE_initial = 1e13 # arbitrary initial mass of MBE in grams
k_values = [1, 2, 4]

plt.figure(figsize=(10, 6))
for k in k_values:
    Gamma_k = Gamma_PBH(k, M_MBE_initial)
    tau_k = tau_PBH(k, M_MBE_initial)
    M_MBE_values = M_MBE_evolve(t, tau_k, Gamma_k, M_MBE_initial, k)
    plt.plot(t, M_MBE_values, label=f"k={k}")

# Plot details
plt.xlabel("Time (s)")
plt.ylabel("M_MBE (grams)")
plt.title("Evolution of M_MBE vs Time for Different k Values")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
G_gray = 3.8
G = 1 / (1.22e19)**2
g_sm = 102.6
M_P = (8 * np.pi * G)**(-0.5)

# Logarithmic time array (scaled to GeV, factor 6.58e-25)
t = np.logspace(1, 100, 1000) * 6.58e-25  # Converted to GeV

# Function definitions
def Gamma_PBH(k, M_MBE):
    return ((G_gray * g_sm / (7680 * np.pi)) * 2**k * (3 + 2*k) * M_P * (M_P / M_MBE)**(3 + 2*k))

def tau_PBH(k, M_MBE):
    return 1 / Gamma_PBH(k, M_MBE)

def M_MBE_evolve(t, tau_PBH, Gamma_PBH_k, M_MBE, k, q=0.5):  # q is an arbitrary small constant
    factor_term = 1 - Gamma_PBH_k * (t - tau_PBH * (1 - q**3))
    factor_term[factor_term <= 0] = np.nan  # Avoid invalid values
    factor = factor_term**(1 / (3 + 2 * k))
    return M_MBE * factor

# Initial parameters and plot setup
M_MBE_initial = 1/2*1e3*5.62*10**23 #GeV  
k_values = [0, 1, 2, 4]
# k_values = [4]

plt.figure(figsize=(10, 6))
for k in k_values:
    Gamma_k = Gamma_PBH(k, M_MBE_initial)
    tau_k = tau_PBH(k, M_MBE_initial)
    print(f"Debug: k={k}, Gamma_k={Gamma_k}, tau_k={tau_k}")  # Debugging info

    # Valid time range (ensuring factor_term > 0)
    valid_t_limit = tau_k * (1 - 0.5**3) / Gamma_k
    valid_t_indices = t <= valid_t_limit
    t_valid = t[valid_t_indices]
    
    if len(t_valid) == 0:
        print(f"No valid time points for k={k}. Skipping.")
        continue

    M_MBE_values = M_MBE_evolve(t_valid, tau_k, Gamma_k, M_MBE_initial, k)
    plt.plot(t_valid, M_MBE_values, label=f"k={k}")

# Plot details
plt.xlabel("Time (GeV^{-1})")
plt.ylabel("M_MBE (GeV)")
plt.xscale("log")
plt.yscale("log")
plt.title("Evolution of M_MBE vs Time for Different k Values")
plt.legend()
plt.grid()
plt.show()


#lifetime close to age of the universe
```
