```python
import numpy as np
import matplotlib.pyplot as plt

# 常數定義
G_gray = 3.8
G = 1 / (1.22e19)**2
g_sm = 102.6
g_sm_SUSY = 200
M_P = (8 * np.pi * G)**(-0.5)
M_MBE_initial = 1.9e14 * 5.62e23  # GeV

# 以 GeV^-1 為單位的對數時間陣列
t = np.logspace(0, 80, 2000000) * 6.58e-25  

# NO SUSY 衰變率
def Gamma_PBH(k, M_MBE):
    return ((G_gray * g_sm / (7680 * np.pi)) * 2**k * (3 + 2*k) * M_P * (M_P / M_MBE)**(3 + 2*k))

# SUSY 衰變率
def Gamma_PBH_SUSY(k, M_MBE):
    return ((G_gray * g_sm_SUSY / (7680 * np.pi)) * 2**k * (3 + 2*k) * M_P * (M_P / M_MBE)**(3 + 2*k))

# 計算黑洞蒸發時間（壽命）
def tau_PBH(k, M_MBE):
    return 1 / Gamma_PBH(k, M_MBE)

def tau_SUSY_PBH(k, M_MBE):
    return 1 / Gamma_PBH_SUSY(k, M_MBE)

# 黑洞質量隨時間演化
def M_MBE_evolve(t, tau_PBH, Gamma_PBH_k, M_MBE, k, q=0.5):
    factor_term = 1 - Gamma_PBH_k * (t)
    #factor_term = 1 - Gamma_PBH_k * (t - tau_PBH * (1 - q**3))
    factor_term[factor_term <= 0] = np.nan  # 避免質量變負
    factor = factor_term**(1 / (3 + 2 * k))
    return M_MBE * factor

# 設定 k 值和顏色對應
k_values = [0]
color_map = {0: "blue", 1: "orange", 2: "green", 4: "red"}

# 畫圖
plt.figure(figsize=(10, 6))

for k in k_values:
    color = color_map[k]  # 取得 k 值的顏色
    
    # NO SUSY
    Gamma_k = Gamma_PBH(k, M_MBE_initial)
    tau_k = tau_PBH(k, M_MBE_initial)
    valid_t_indices = t <= tau_k
    t_valid = t[valid_t_indices]
    
    if len(t_valid) > 0:
        M_MBE_values = M_MBE_evolve(t_valid, tau_k, Gamma_k, M_MBE_initial, k)
        plt.plot(t_valid, M_MBE_values, linestyle="-", linewidth=1.5, color=color, label=f"NO SUSY k={k}")

    # SUSY
    Gamma_k_SUSY = Gamma_PBH_SUSY(k, M_MBE_initial)
    tau_k_SUSY = tau_SUSY_PBH(k, M_MBE_initial)
    valid_t_indices_SUSY = t <= tau_k_SUSY
    t_valid_SUSY = t[valid_t_indices_SUSY]

    if len(t_valid_SUSY) > 0:
        M_MBE_values_SUSY = M_MBE_evolve(t_valid_SUSY, tau_k_SUSY, Gamma_k_SUSY, M_MBE_initial, k)
        plt.plot(t_valid_SUSY, M_MBE_values_SUSY, linestyle="--", linewidth=1.5, color=color, label=f"SUSY k={k}")
        
        
        
plt.axhline(y=0.5 * M_MBE_initial, color="purple", linestyle="--", label="Half of $M_{initial}$")

# 定義目標值
target_mass = 0.5 * M_MBE_initial

# 找到 M_MBE_values 與 target_mass 的最近索引
diff = np.abs(M_MBE_values - target_mass)
min_index = np.argmin(diff)  # 找到最接近目標質量的索引

# 獲得對應時間
intersection_time = t_valid[min_index]

print(f"交點時間 (GeV^-1): {intersection_time}")
print(f"交點質量 (GeV): {M_MBE_values[min_index] * 1.78 * 10**(-24)}")
        

plt.scatter(intersection_time, target_mass, color="red", marker="o")

# 找到 M_MBE_values 與 target_mass 的最近索引
diff = np.abs(M_MBE_values_SUSY - target_mass)
min_index = np.argmin(diff)  # 找到最接近目標質量的索引

# 獲得對應時間
intersection_time = t_valid[min_index]

print(f"交點時間SUSY (GeV^-1): {intersection_time}")
print(f"交點質量SUSY (GeV): {M_MBE_values[min_index] * 1.78 * 10**(-24)}")
        

plt.scatter(intersection_time, target_mass, color="blue", marker="o")


# 畫出目前宇宙年齡的參考線
current_universe_age = 6.6 * 1e41  # GeV^-1
plt.axvline(x=current_universe_age, color="black", linestyle=":", label="Current Universe Age")




# 找到藍色曲線與目前宇宙年齡（黑色虛線）的交點 #SUSY
diff_SUSY = np.abs(t_valid_SUSY - current_universe_age)  # 計算時間差
min_index_SUSY = np.argmin(diff_SUSY)  # 找到最接近目前宇宙年齡的索引

# 獲得對應質量
intersection_time_SUSY = t_valid_SUSY[min_index_SUSY]
intersection_mass_SUSY = M_MBE_values_SUSY[min_index_SUSY]

print(f"交點時間SUSY (GeV^-1): {intersection_time_SUSY}")
print(f"交點質量SUSY (GeV): {intersection_mass_SUSY * 1.78 * 10**(-24)}")

# 在圖表中標註交點
plt.scatter(intersection_time_SUSY, intersection_mass_SUSY, color="green", marker="x", label="Current mass of PBH with SUSY")

# 顯示交點的座標
plt.text(intersection_time_SUSY, intersection_mass_SUSY, 
         f"({intersection_time_SUSY:.2e}, {intersection_mass_SUSY * 1.78 * 10**(-24):.2e})", 
         color="green", fontsize=15, verticalalignment='bottom', horizontalalignment='right')

# 找到 NO SUSY 曲線與目前宇宙年齡（黑色虛線）的交點
diff_NO_SUSY = np.abs(t_valid - current_universe_age)  # 計算時間差
min_index_NO_SUSY = np.argmin(diff_NO_SUSY)  # 找到最接近目前宇宙年齡的索引

# 獲得對應質量
intersection_time_NO_SUSY = t_valid[min_index_NO_SUSY]
intersection_mass_NO_SUSY = M_MBE_values[min_index_NO_SUSY]

print(f"交點時間NO SUSY (GeV^-1): {intersection_time_NO_SUSY}")
print(f"交點質量NO SUSY (GeV): {intersection_mass_NO_SUSY * 1.78 * 10**(-24)}")

# 在圖表中標註交點
plt.scatter(intersection_time_NO_SUSY, intersection_mass_NO_SUSY, color="red", marker="x", label="Current mass of PBH")

# 顯示交點的座標
plt.text(intersection_time_NO_SUSY, intersection_mass_NO_SUSY, 
         f"({intersection_time_NO_SUSY:.2e}, {intersection_mass_NO_SUSY * 1.78 * 10**(-24):.2e})", 
         color="red", fontsize=15, verticalalignment='bottom', horizontalalignment='right')





plt.tick_params(axis='both', which='major', labelsize=18)  # 調整主要刻度字體大小
plt.tick_params(axis='both', which='minor', labelsize=14)  # 調整次要刻度字體大小

plt.xlim(1e14, 1e44)
plt.ylim(2e36, 1e38)

# 圖表細節
plt.xlabel("Time ($GeV^{-1}$)",fontsize = 15)
plt.ylabel("$M_{MBE}$ (GeV)",fontsize = 15)
plt.xscale("log")
plt.yscale("log")

value = M_MBE_initial / 5.62e23
power = int(np.floor(np.log10(value)))  # 計算次方部分
mantissa = value / (10**power)  # 計算尾數部分

plt.title(
    fr"Evolution of $M_{{MBE}}$ vs Time (no SUSY vs SUSY), $M_{{\text{{initial}}}} = {mantissa:.2f} \times 10^{{{power}}}g$",
    fontsize=15
)

plt.legend(fontsize = 15)
plt.grid()
plt.show()




print(tau_PBH(0, M_MBE_initial))
print(tau_SUSY_PBH(0, M_MBE_initial))
print(tau_PBH(1, M_MBE_initial))
print(tau_SUSY_PBH(1, M_MBE_initial))
print(tau_PBH(2, M_MBE_initial))
print(tau_SUSY_PBH(2, M_MBE_initial))
print(tau_PBH(4, M_MBE_initial))
print(tau_SUSY_PBH(4, M_MBE_initial))



###no susy 掉到 1/10mass###
####字大一點#####
```
