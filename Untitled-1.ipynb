```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
G_gray = 3.8
G = 1 / (1.22e19)**2
g_sm = 102.6
M_P = (8 * np.pi * G)**(-0.5)

def Gamma_PBH(k, M_MBE):
    return (G_gray * g_sm / (7680 * np.pi) * 2**k * (3 + 2*k) * M_P * (M_P / M_MBE)**(3 + 2*k))

def tau_PBH(k, M_MBE):
    return 1 / Gamma_PBH(k, M_MBE)

def M_MBE_evolve(t, tau_PBH, Gamma_PBH_k, M_MBE, k, q=0.5):
    factor = (1 - Gamma_PBH_k * (t - tau_PBH * (1 - q**3)))**(1 / (3 + 2 * k))
    return M_MBE * factor

# Filepath for the data
file_path = "/home/leah/Downloads/blackhawk_v2.3/results/test_no_MBE/life_evolutions.txt"

# Reading the data, skipping the header lines
data = np.genfromtxt(file_path, skip_header=3, invalid_raise=False)  # Skips the first 3 lines
time = data[:, 0]  # First column is time
mass = data[:, 1]  # Second column is mass

# Convert mass to GeV
mass_gev = mass * 5.62e23

# Find the index where mass <= half of its initial value
half_mass = mass_gev[0] / 2
if np.any(mass_gev <= half_mass):
    half_index = np.where(mass_gev <= half_mass)[0][0]
else:
    print("Warning: Mass does not decay to half its initial value within the dataset.")
    half_index = len(mass_gev) - 1

# Trim time and mass arrays
time_gev = time / 6.58e-25  # Convert time to GeV
time_gev_trimmed = time_gev[:half_index + 1]
mass_gev_trimmed = mass_gev[:half_index + 1]

# Calculate dM/dt for the trimmed data
dt = np.diff(time[:half_index + 1])
dM = np.diff(mass_gev[:half_index + 1])
dM_dt = dM / dt
time_mid = (time[:-1] + time[1:])[:half_index] / 2

# Set up initial conditions for ""
t = np.logspace(np.log10(time_gev_trimmed[-1]), 41, 1000)  # Time in GeV
# M_MBE_initial = mass_gev_trimmed[-1]
M_MBE_initial = 2.81*1e32
k_values = [1, 2, 4]

# Plot combined figure
fig, ax1 = plt.subplots(figsize=(12, 8))

# Plot the "" data
ax1.plot(time_gev_trimmed, mass_gev_trimmed, label="Black Hole Mass (GeV)", color="blue", lw=2)
ax1.set_xlabel("Time (GeV^{-1})", fontsize=12)
ax1.set_ylabel("Mass (GeV)", fontsize=12, color="blue")
ax1.tick_params(axis='y', labelcolor="blue")
ax1.set_xscale("log")
ax1.set_yscale("log")

# Add dM/dt plot
ax2 = ax1.twinx()
ax2.plot(time_mid[:half_index] * 6.58e-25, dM_dt, label="dM/dt (GeV/s)", color="red", lw=2, linestyle="--")
ax2.set_ylabel("dM/dt (GeV/s)", fontsize=12, color="red")
ax2.tick_params(axis='y', labelcolor="red")
ax2.set_yscale("log")

# Plot the "" M_MBE evolution
for k in k_values:
    Gamma_k = Gamma_PBH(k, M_MBE_initial)
    tau_k = tau_PBH(k, M_MBE_initial)
    M_MBE_values = M_MBE_evolve(t, tau_k, Gamma_k, M_MBE_initial, k)
    ax1.plot(t, M_MBE_values, label=f"M_MBE (k={k})", linestyle="-", lw=2)

# Add legends and finalize plot
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc="upper right")
ax1.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.title("Combined Evolution of Black Hole Mass and M_MBE over Time", fontsize=14)
plt.tight_layout()
plt.show()
```
