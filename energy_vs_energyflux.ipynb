```python
# this is test!!!!!!!!!

import numpy as np
import matplotlib.pyplot as plt

# Load the neutrino spectrum data, skipping header lines or comments
data = np.loadtxt('./neutrinos_primary_spectrum.txt', comments='#', skiprows=2)

# Extract energy and flux columns
energy = data[:, 0]  # Energy in GeV
flux = data[:, 1]    # Flux in GeV^-1 cm^-2 s^-1

# Compute E^2 * Phi_nu
e2_flux = energy**2 * flux

# Plot E^2 * Phi_nu
plt.figure(figsize=(8, 6))
plt.loglog(energy, e2_flux, label=r"$E^2 \Phi_\nu$")
plt.xlabel(r"Energy $E$ [GeV]")
plt.ylabel(r"$E^2 \Phi_\nu$ [GeV cm$^{-2}$ s$^{-1}$]")
plt.title("Neutrino Energy Flux")
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()
```
